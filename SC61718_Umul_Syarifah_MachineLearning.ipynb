## Menangani Outlier
Penanganan outlier hanya bisa dilakukan pada tipe data kontinu atau numerik bukan data kategorikal

````python
kolom_heart1 = ['age','chol','trestbps','thalach','oldpeak']

batas_kolom = {}
````

````python
for kolom in kolom_heart1:
    Q1 = heart_df_nodup[kolom].quantile(0.25)
    Q3 = heart_df_nodup[kolom].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    batas_kolom[kolom] = (lower_bound, upper_bound)

outliers = {}
for kolom in kolom_heart1:
    lower_bound, upper_bound = batas_kolom[kolom]
    outliers[kolom] = heart_df_nodup[(heart_df_nodup[kolom] < lower_bound) | (heart_df_nodup[kolom] > upper_bound)]
````

````python
for kolom, (lower_bound, upper_bound) in batas_kolom.items():
    print(f"Kolom '{kolom}':")
    print(f"  Lower Bound: {lower_bound}")
    print(f"  Upper Bound: {upper_bound}")
    print()
````

````python
# Inisialisasi DataFrame baru untuk menyimpan data selain outlier
df_no_outliers = heart_df_nodup.copy()

# Loop melalui setiap kolom dan hapus baris yang merupakan outlier
for kolom in kolom_heart1:
    lower_bound, upper_bound = batas_kolom[kolom]
    df_no_outliers = df_no_outliers[~((df_no_outliers[kolom] < lower_bound) | (df_no_outliers[kolom] > upper_bound))].reset_index(drop=True)

# Cetak DataFrame baru yang hanya berisi data selain outlier
print("DataFrame baru tanpa outlier:")
df_no_outliers
````

### Cross Validation
cross validation adalah metode evaluasi model ML untuk menguji efektivitas suatu model dengan melibatkan pelatihan menggunakan sebagian data input dan pengujian pada sebagian data input yang berbeda. ### Learning Curve
Learning curve atau kurva pembelajaran adalah kurva untuk mengevaluasi model machine learning yang telah dibuat untuk mengetahui hasil fitting apakah overfit, underfit, atau good fit. Hal ini perlu dilakukan supaya kita bisa mengetahui langkah selanjutnya untuk membuat model machine learning dengan performa yang baik.

- Overfitting : terjadi jika hasil akurasi train set lebih tinggi dibandingkan test set. Saat training hasilnya akurasinya sangat tinggi tetapi setelah dievaluasi dengan test set hasil akurasinya jauh lebih rendah.
- Underfitting : terjadi jika hasil training dan test set sama-sama menmiliki akurasi yang rendah dalam memprediksi suatu data
- Good fit : terjadi jika hasil training pada set training dan testing memiliki akurasi yang tinggi.

````python
#set train size
#316 x 80% = 252 (20% for validation)
train_sizes = [1, 20, 40, 50, 100, 150, 252]
````

````python
from sklearn.model_selection import learning_curve
def learn_curve_plot(estimator, x, y, cv, train_sizes):
  train_sizes, train_scores, validation_scores = learning_curve(estimator,X=X,y=y, train_sizes= train_sizes, cv=cv)
  train_scores_mean = train_scores.mean(axis=1)
  validation_scores_mean = validation_scores.mean(axis=1)

  #Print
  print('Mean training scores\n\n', pd.Series(train_scores_mean, index=train_sizes))
  print('\n', '-' * 20) # separator
  print('\nMean validation scores\n\n',pd.Series(validation_scores_mean, index=train_sizes))

  plt.style.use('seaborn')
  plt.plot(train_sizes, train_scores_mean, label = 'Training error')
  plt.plot(train_sizes, validation_scores_mean, label = 'Validation error')
  plt.ylabel('Accuracy', fontsize = 14)
  plt.xlabel('Training set size', fontsize = 14)
  plt.title('Learning Curve', fontsize = 18, y = 1.03)
  plt.legend()
````

````python
# Learning Curve Logistic Regression
learn_curve_plot(estimator=model1, x=X, y=y, cv=5, train_sizes=train_sizes)
````

Perlu dilakukan variasi data saat training, menggunakan sedikit fitur, dan regulasi. Dalam kasus daatset ini, perlu ditinjau fitur mana yang paling penting terhadap target.

### Before using Cross Validation

````python
models = {"Logistic Regression": linear_model.LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier(),
          "Decission Tree" :DecisionTreeClassifier(),
          "Gradient Boosting" : GradientBoostingClassifier(),
          "SVM" : SVC()}

# function to fit and score models

def fit_and_score(models, X_train, X_test, y_train, y_test):
    np.random.seed(42)
    model_scores = {}
    for name, model in models.items():
        #fit the model
        model.fit(X_train, y_train)
        #append the score of the model
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
````
